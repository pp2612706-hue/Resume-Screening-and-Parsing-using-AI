```python
def extract_skills(text):
    match = re.search(r"skills in (.*?)[.] He has|skills in (.*?)[.] She has", text, re.IGNORECASE)
    if match:
        return match.group(1) if match.group(1) else match.group(2)
    return ""

df["skills_extracted"] = df["resume_text"].apply(extract_skills)
```

```python
def extract_experience(text):
    match = re.search(r"(\d+)\s+years of experience", text, re.IGNORECASE)
    return int(match.group(1)) if match else 0

df["experience_extracted"] = df["resume_text"].apply(extract_experience)
```

```python
def extract_experience(text):
    match = re.search(r"(\d+)\s+years of experience", text, re.IGNORECASE)
    return int(match.group(1)) if match else 0

df["experience_extracted"] = df["resume_text"].apply(extract_experience)
```

```python
def extract_education(text):
    match = re.search(r"Education:\s*([A-Za-z. ]+)", text)
    return match.group(1).strip() if match else ""

df["education_extracted"] = df["resume_text"].apply(extract_education)
```

```python
def extract_certifications(text):
    match = re.search(r"Certifications:\s*(\d+)", text)
    return int(match.group(1)) if match else 0

df["certifications_extracted"] = df["resume_text"].apply(extract_certifications)
```

```python
job_description = "python sql machine learning nlp communication recruitment analytics power bi"
```

```python
required_keywords = ["python", "sql", "machine learning", "nlp", "communication", "recruitment", "analytics", "power bi"]

def keyword_match_count(skill_text):
    skill_text = skill_text.lower()
    return sum(1 for k in required_keywords if k in skill_text)

df["keyword_match_count"] = df["skills_extracted"].apply(keyword_match_count)
```

```python
documents = [job_description] + df["skills_extracted"].tolist()

vectorizer = TfidfVectorizer()
tfidf_matrix = vectorizer.fit_transform(documents)
```

```python
similarity_scores = cosine_similarity(tfidf_matrix[0:1], tfidf_matrix[1:]).flatten()
df["similarity_score"] = similarity_scores
```

```python
df["experience_score"] = df["experience_extracted"] / df["experience_extracted"].max()
df["certification_score"] = df["certifications_extracted"] / df["certifications_extracted"].max()
```

```python
education_map = {
    "B.Com": 0.50, "BBA": 0.55, "BCA": 0.55, "B.Sc": 0.60,
    "B.Tech": 0.70, "MBA": 0.75, "MBA HR": 0.80,
    "MCA": 0.80, "M.Sc": 0.80, "M.Tech": 0.90, "M.Stat": 0.90
}

df["education_score"] = df["education_extracted"].map(education_map)
```

```python
df["keyword_score"] = df["keyword_match_count"] / len(required_keywords)

df["final_score"] = (
    0.30 * df["keyword_score"] +
    0.30 * df["similarity_score"] +
    0.20 * df["experience_score"] +
    0.10 * df["education_score"] +
    0.10 * df["certification_score"]
)
```

```python
ranked_df = df.sort_values(by="final_score", ascending=False).reset_index(drop=True)
ranked_df.head()
```

```python
top_5 = ranked_df.head(5)
top_5
```

```python
def label_candidate(score):
    if score >= 0.65:
        return "Strong Fit"
    elif score >= 0.50:
        return "Moderate Fit"
    else:
        return "Low Fit"

ranked_df["recommendation"] = ranked_df["final_score"].apply(label_candidate)
```

```python
final_output = ranked_df[[
    "candidate_id",
    "resume_text",
    "skills_extracted",
    "experience_extracted",
    "education_extracted",
    "certifications_extracted",
    "keyword_match_count",
    "keyword_score",
    "similarity_score",
    "experience_score",
    "education_score",
    "certification_score",
    "final_score",
    "recommendation"
]]

final_output
```
